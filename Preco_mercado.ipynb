```python
import yfinance as yf
import pandas as pd
import os

# Dicionário com o nome da empresa e o ticker (código do ativo no Yahoo Finance)
ativos = {
    'Google': 'GOOGL',
    'Amazon': 'AMZN',
    'Microsoft': 'MSFT',
    'Bitcoin': 'BTC-USD',
    'Mercado Livre': 'MELI'
}

# Criar uma pasta para salvar os arquivos
os.makedirs('dados_ativos', exist_ok=True)

for nome, ticker in ativos.items():
    # Baixando o histórico do último ano
    df = yf.download(ticker, period="1y")

    # Resetando o index para transformar a Data em coluna
    df = df.reset_index()

    # Criando a coluna com o Código do Ativo para facilitar no Power BI
    df['Codigo_Ativo'] = ticker

    # Selecionando apenas as colunas relevantes (Data, Fechamento, etc)
    df = df[['Date', 'Close', 'Codigo_Ativo']]
    df.columns = ['Data', 'Preco_Fechamento', 'Codigo_Ativo'] # Renomeando

    # Exportando para CSV
    nome_arquivo = f"dados_ativos/{ticker}_historico.csv"
    df.to_csv(nome_arquivo, index=False)

print("Arquivos CSV gerados com sucesso na pasta 'dados_ativos'!")
```

```python

#criar variavel para receber os dados das tabela. Essas variaveis vão receber o resultado
# da funcao do pandas que le um arquivo .csv

#as variaveis vendas_janeiro, vendas_fevereiro e vendas_marco ...
#estão recebendo o conteudo dos arquvivos .csv que estão sendo lidos pela função read da biblioteca pandas


amzn = pd.read_csv('/content/dados_ativos/AMZN_historico.csv')
btc = pd.read_csv('/content/dados_ativos/BTC-USD_historico.csv')
googl = pd.read_csv('/content/dados_ativos/GOOGL_historico.csv')
meli = pd.read_csv('/content/dados_ativos/MELI_historico.csv')
msft = pd.read_csv('/content/dados_ativos/MSFT_historico.csv')


```

```python

#para transformar os dados destres arquivos .csv em dataframe usando esse função da biblioteca panda
# ex df_dados = pd.DataFrame(dados)



df_amzn = pd.DataFrame(amzn)
df_btc = pd.DataFrame(btc)
df_googl = pd.DataFrame(googl)
df_meli = pd.DataFrame(meli)
df_msft = pd.DataFrame(msft)
```

```python
#usa a funcao pd.concat para juntar os dados de tabelas diferentes
# pd.concat([dados_antigos, dados_novos], ignore_index=True)

df_preco_mercado = pd.concat([df_amzn,df_btc,df_googl,df_meli,df_msft], ignore_index=True)

display(df_preco_mercado)
```

```python
#para exportar o arquivo .csv para seu pc
df_preco_mercado.to_csv("Preço_mercado.csv", index=False)
```
